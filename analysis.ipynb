Befor we delve into the data here is a quick overview of what we are going to be dealing with. 

This is a historical dataset on the modern Olympic Games, including all the Games from Athens 1896 to Rio 2016. It includes 15 features as show below. 

`ID (int64):` Unique number for each athlete.

`Name (object):`  Athlete's name.

`Sex (object):` Sex of athete M: male or W:women

`Age (float64):` Age of athlete.

`Height (float64 - cm):` Height of athlete in cm.

`Weight (float64 - kg):` Weight of athlete in cm.

`Team (object):` Team country of athelete. 

`NOC (object):` National Olympic Committee by each country.

`Games (object):` Year and season of the Olyimic Game.

`Year (int64):` Year the athlete competed.

`Season (object):` Season the Olymics were played. 

`City (object):` The city where the Olymics were hosted. 

`Sport (object):` The sport the athelete competed in. 

`Event (object):` The name of the event the athelete competed in. 

`Medal (object):` The medal the athelete won in the competition Gold, Silver, Bronze, or NA.

Lets check the unique values in each feature. 

```python
column = df.columns.to_list()

for column_name in column:
    df[column_name].unique()
    print(f"The column: {column_name} has {len(df[column_name].unique())} unique values that are \n {df[column_name].unique()}\n")

```
